```python
#Bioinformatics Example: DNA Sequence Matching

def sequence_match(seq1, seq2):
    match = 0
    length = min(len(seq1), len(seq2))

    for i in range(length):
        if seq1[i] == seq2[i]:
            match += 1

    similarity = (match / length) * 100
    return similarity


# Input DNA sequences
seq1 = "ATGCTAGCTAG"
seq2 = "ATGCGAGCTAG"

# Calculate similarity
result = sequence_match(seq1, seq2)

print("Sequence 1:", seq1)
print("Sequence 2:", seq2)
print("Similarity:", result, "%")
```
